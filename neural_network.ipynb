```python
from sklearn.datasets import make_moons
```

```python
import torch
```

```python
import torch.nn as nn
```

```python
import torch.optim as optim
```

```python
X,Y = make_moons(n_samples=1000, noise=0.15, random_state=0)
```

```python
model = nn.Sequential(
    nn.Linear(2,1000),
    nn.ReLU(),
    nn.Linear(1000,2),
    # nn.Softmax(dim=1)
)
```

```python
model
```

```python
X = torch.FloatTensor(X)
Y = torch.LongTensor(Y)
```

```python
# lsfn = nn.NLLLoss()
lossfn = nn.CrossEntropyLoss()
sgd_opt = optim.SGD(model.parameters(), lr=0.1)
```

```python
for _ in range(20):
    sgd_opt.zero_grad()
    Yp = model(X)
    loss = lossfn(Yp, Y)
    loss.backward()
    sgd_opt.step()

    print(loss.item())
```

```python
Ypred = model(X)
```

```python
softout = nn.Softmax(dim=1)(Ypred)
```

```python
torch.argmax(softout, dim=1)
```

```python
import numpy as np
import pandas as pd
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
dt = load_iris()
```

```python
dt.keys()
```

```python
X, Y = dt.data, dt.target
```

```python
dt.target_names
```

```python
model = nn.Sequential(
    nn.Linear(4, 30),
    nn.ReLU(),
    nn.Linear(30, 10),
    nn.ReLU(),
    nn.Linear(10, 3),
    # nn.Softmax(dim=1)
)
```

```python
X = torch.FloatTensor(X)
Y = torch.LongTensor(Y)
```

```python
lossfn = nn.CrossEntropyLoss()
optSgd = optim.SGD(model.parameters(), lr=0.01)
```

```python
for _ in range(200):
    optSgd.zero_grad()
    Yp = model(X)
    loss = lossfn(Yp, Y)
    loss.backward()
    optSgd.step()

    print(loss.item())
```

```python
Ynewp = model(X)
```

```python
sftout = torch.softmax(Ynewp, dim=1)
```

```python
torch.argmax(sftout, dim=1)
```

```python
from torch.utils.data import DataLoader
```

```python
df = pd.DataFrame(X)
```

```python
df['Y'] = Y
```

```python
data = torch.tensor(df.to_numpy())
```

```python
modl = nn.Sequential(
    nn.Linear(4, 30),
    nn.ReLU(),
    nn.Linear(30, 10),
    nn.ReLU(),
    nn.Linear(10, 3),
    # nn.Softmax(dim=1)
)
```

```python
lossfn = nn.CrossEntropyLoss()
batchSgd = optim.SGD(modl.parameters(), lr=0.01)
```

```python
batches = DataLoader(data, batch_size=10, shuffle=True)
```

```python
modl.train()
for epoch in range(200):
    batch_loss = 0
    for i, batch in enumerate(batches):
        batch_X = batch[:, :-1].float()
        batch_Y = batch[:, -1].long()

        batchSgd.zero_grad()
        batch_Yp = modl(batch_X)
        loss = lossfn(batch_Yp, batch_Y)
        loss.backward()
        batchSgd.step()

        batch_loss = batch_loss + loss.item()
    print(f"Epoch {epoch+1}: Total Loss = {batch_loss}")
```
